```python
import random
import matplotlib.pyplot as plt
```

```python
input_array = [x for x in range(50)]
```

```python
output_array = [3*x for x in range(50)]
#print(output_array)
```

```python
randomized_level = 5
noisy_array = []
#Add noise to the output array
for number in output_array:
    number = number + random.uniform(-randomized_level, randomized_level)
    noisy_array.append(number)
    pass
#print(noisy_array)
```

```python
plt.plot(output_array)
plt.plot(noisy_array, 'x')
plt.show()
```

```python
import tensorflow as tf
import numpy as np
```

```python
print(input_array)
print(noisy_array)

output_array = np.array(output_array)
output_array = np.reshape(output_array, (50,1))

input_array = np.array(input_array)
input_array = np.reshape(input_array, (50,1))

noisy_array = np.array(noisy_array)
noisy_array = np.reshape(noisy_array, (50,1))
```

```python
print(input_array.shape)
print(output_array.shape)
print(noisy_array.shape)
```

```python
layer1 = tf.keras.layers.Input(shape=1)
layerHidden = tf.keras.layers.Dense(units=1)(layer1)
myModel = tf.keras.Model(inputs=layer1, outputs=layerHidden)
myModel.summary()
```

```python
myModel.compile(loss='mae', optimizer=tf.keras.optimizers.RMSprop(learning_rate=0.01), metrics=["accuracy"])
```

```python
myModel.fit(x=input_array, y=noisy_array, epochs=500)
```

```python
myVar = tf.Variable(1337, dtype="float32")
myVar = tf.expand_dims(myVar, 0)
prediction = myModel(myVar)
print(prediction)
```
